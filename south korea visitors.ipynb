data  visualization

```python
plt.figure(figsize=(18,10))
plt.title('visitor by year',fontsize='20')
plt.xlabel('year',fontsize='20')
plt.ylabel('visitor',fontsize='20')
plt.bar(korea_df.date,korea_df.visitor)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='tourism',y='date',data=korea_df)
```

```python
plt.figure(figsize=(12,6))
plt.title('studying by year')
plt.xlabel('year')
plt.ylabel('studying')
plt.plot(korea_df.date,korea_df.studying)
```

```python
plt.figure(figsize=(12,6))
sns.set_style('darkgrid')
plt.title('business &official affairs by year')
sns.lineplot(x='date',y='business',data=korea_df)
sns.lineplot(x='date',y='official affairs',data=korea_df)
plt.legend(['business','official affairs'])
```

```python
top=korea_df.nation.head()
top
```

```python
for country in top:
    plt.figure(figsize=(12,6))
    plt.plot(korea_df[korea_df['nation']==country].date,korea_df[korea_df['nation']==country].visitor)
    plt.title('visitor of {} by date'.format(country))
    plt.xlabel('date',fontsize=15)
    plt.ylabel('visitor',fontsize=15)
```

```python
less=korea_df.nation.tail()
less
```

```python
for country in less:
    plt.figure(figsize=(10,7))
    plt.bar(korea_df[korea_df['nation']==country].date,korea_df[korea_df['nation']==country].visitor)
    plt.title('visitor of {} by date'.format(country))
    plt.xlabel('date',fontsize=12)
    plt.ylabel('visitor',fontsize=12)
    
```

```python
plt.figure(figsize=(10,5))
plt.plot(korea_df[korea_df['nation']=='USA'].date,korea_df[korea_df['nation']=='USA'].studying,c='r',ls='-',marker='s')
plt.plot(korea_df[korea_df['nation']=='China'].date,korea_df[korea_df['nation']=='China'].studying,'o--g')
plt.title('USA vs China')
plt.xlabel('date',fontsize=10)
plt.ylabel('studying',fontsize=10)
plt.legend(['USA','China'])
```

```python
plt.figure(figsize=(20,16))
sns.lineplot(x='date',
             y='visitor',
             hue='nation',
             data=korea_df)
plt.legend(loc='center left',bbox_to_anchor=(1,0.5))
plt.title('visitors of all country',fontsize=20)             
             
            
```

```python
plt.figure(figsize=(20,16))
sns.lineplot(x='date', y='share',hue='nation',data=korea_df)
             
plt.legend(loc='center left',bbox_to_anchor=(1,0.5))
plt.title('Percentage of all visitors in the month for every country',fontsize=20)
```

```python
plt.figure(figsize=(20,16))
sns.lineplot(x='date', y='growth',hue='nation',data=korea_df)
             
plt.legend(loc='center left',bbox_to_anchor=(1,0.5))
plt.title(' Growth percentage in the number of visitors compared to the same month last year for every country',fontsize=20)
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='date',
                y='business',
                data=korea_df,
                hue='male',
                s=30) 
               

plt.title('men in business',fontsize=15)
```

```python
plt.title('distribution of crewman',fontsize=12)
plt.hist(korea_df.crewman);
```

```python
plt.figure(figsize=(12,6))

plt.title('distribution of visitor',fontsize=12)

plt.hist([korea_df.tourism, korea_df.studying, korea_df.business,korea_df.others],
         bins=np.arange(2,10, 0.25),stacked=True) 
plt.legend(['tourism','business','studying','others'])
```
